```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img= cv2.imread("sugar-coated-chocolate-gems-candy-260nw-1544666168.jpg", cv2.IMREAD_GRAYSCALE)

images=[img]

for i in range(1):
    plt.subplot(1,1,i+1), plt.imshow(images[i], 'gray')
    plt.title('Images')
    plt.xticks([]), plt.yticks([])


plt.show()
```

```python
img= cv2.imread("sugar-coated-chocolate-gems-candy-260nw-1544666168.jpg", cv2.IMREAD_GRAYSCALE)
_, mask= cv2.threshold(img, 120,255, cv2.THRESH_BINARY_INV)

images=[img, mask]
titles=['Image', 'Mask']

for i in range(2):
    plt.subplot(1,2,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])


plt.show()
```

```python
img= cv2.imread("sugar-coated-chocolate-gems-candy-260nw-1544666168.jpg", cv2.IMREAD_GRAYSCALE)
_, mask= cv2.threshold(img, 120,255, cv2.THRESH_BINARY_INV)
kernel=np.ones((2,2), np.uint8)
dilation= cv2.dilate(mask, kernel, iterations=3)

images=[img, mask,dilation]
titles=['Image', 'Mask','dilation']

for i in range(3):
    plt.subplot(1,3,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])


plt.show()
```

```python
img= cv2.imread("sugar-coated-chocolate-gems-candy-260nw-1544666168.jpg", cv2.IMREAD_GRAYSCALE)
_, mask= cv2.threshold(img, 100,255, cv2.THRESH_BINARY_INV)
kernel=np.ones((2,2), np.uint8)
dilation= cv2.dilate(mask, kernel, iterations=3)
erosion= cv2.erode(mask, kernel, iterations=1)
images=[img, mask,dilation, erosion]
titles=['Image', 'Mask','dilation', 'Erosion']

for i in range(4):
    plt.subplot(2,2,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])


plt.show()
```

```python
img= cv2.imread("sugar-coated-chocolate-gems-candy-260nw-1544666168.jpg", cv2.IMREAD_GRAYSCALE)
_, mask= cv2.threshold(img, 100,255, cv2.THRESH_BINARY_INV)
kernel=np.ones((5,5), np.uint8)
dilation= cv2.dilate(mask, kernel, iterations=3)
erosion= cv2.erode(mask, kernel, iterations=1)
opening= cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel)
closing= cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel)
grad= cv2.morphologyEx(mask, cv2.MORPH_GRADIENT, kernel)
top_hat= cv2.morphologyEx(mask, cv2.MORPH_TOPHAT, kernel)

images=[img, mask,dilation, erosion, opening, closing, grad,top_hat]
titles=['Image', 'Mask','dilation', 'Erosion',"MOrphologycal", 'closing','gardient','TOP_HAT']

for i in range(8):
    plt.subplot(2,4,i+1), plt.imshow(images[i], 'gray')
    plt.title(titles[i])
    plt.xticks([]), plt.yticks([])


plt.show()
```
