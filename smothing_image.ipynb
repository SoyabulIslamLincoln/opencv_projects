```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#img=cv2.imread('c41cd782b962e5ba758c9bb7104b9400.jpg')
img=cv2.imread('sun-hing-dim-sum-basket-dim-sum-1700x1133-1024x682.jpg')
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
kernel=np.ones((5,5), np.float32)/25 
dst=cv2.filter2D(img, -1, kernel)
blur=cv2.blur(img, (5,5))
gaussian= cv2.GaussianBlur(img, (5,5),0)

title_m=['Image',"2d Comvolution", 'blur',"Gaussian"]
image=[img,dst,blur,gaussian]

for i in range(4):
    
    plt.subplot(2,2,i+1), 
    plt.imshow(image[i], 'gray')
    #plt.imshow(image[i], c_map='hot')
    plt.title(title_m[i])
    plt.xticks([]), plt.yticks([])
    

plt.show()
```

```python
img=cv2.imread('Two-examples-of-noisy-images-salt-and-pepper-noise-on-the-left-and-Gaussian-noise.png')
#img=cv2.imread('index.jpeg')
img=cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
kernel=np.ones((5,5), np.float32)/25 
dst=cv2.filter2D(img, -1, kernel)
blur=cv2.blur(img, (5,5))
gaussian= cv2.GaussianBlur(img, (5,5),0)
median=cv2.medianBlur(img, 5)
bilateral= cv2.bilateralFilter(img,9,75,75)

title_m=['Image',"2d Comvolution", 'blur',"Gaussian", 'Median', "Bilateral"]
image=[img,dst,blur,gaussian, median, bilateral]

for i in range(6):
    
    plt.subplot(3,2,i+1), 
    plt.imshow(image[i], 'gray')
    #plt.imshow(image[i], c_map='hot')
    plt.title(title_m[i])
    plt.xticks([]), plt.yticks([])
    

plt.show()
```
